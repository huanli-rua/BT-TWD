在运行前，请确保对应配置文件路径有效，且所需依赖（如 `matplotlib`、`scikit-learn`、`pandas` 等）已安装。

```python
from bttwdlib import visualize_fallback_with_tsne
from IPython.display import Image

# 设置配置文件路径
config_path = "configs/adult_bttwd.yaml"  # 配置文件路径

# 调用 t-SNE 可视化函数
results = visualize_fallback_with_tsne(config_path=config_path)

# 显示保存的图片
Image(filename=results["figure_path"])  # 在 Notebook 中显示可视化图片

# 输出其他结果路径
print(f"t-SNE 嵌入结果保存路径：{results['embedding_path']}")
print(f"弱桶对比摘要保存路径：{results['summary_path']}")
print(f"可视化图片保存路径：{results['figure_path']}")
```
